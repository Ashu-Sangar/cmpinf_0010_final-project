Let's get started by loading in the toxic release data with pandas.

```python
import pandas as pd

water2_data = pd.read_csv("TRIwater2.csv")
water2_data.head(5)
```

The dataframe is ordered by each individual case of toxic material released. That is not really helpful by itself. We went to know where these releases happened. Lets try to sort by adress, since there is no column for neighborhood data. Lets get the case counts for each adress.

```python
water2_data['STREET_ADDRESS'].value_counts()
```

Now lets graph it to see what locations are cases most prevelant.

```python
water2_data['STREET_ADDRESS'].value_counts().plot(kind='bar', figsize=(20, 10))
```

Now that we have the adresses and the number of cases of toxic material releases, we can see which neighborhoods have the most and which have the least. Using google maps, I found all the adresses and sorted them into each of their respective neighborhoods.

```python
sites_data = pd.read_csv("Neighborhood_sites.csv")
sites_data.tail(5)
```
